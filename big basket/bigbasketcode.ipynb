```python
def GetProductData(i):
    sleep(4)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')
    type(soup)
    bs4.BeautifulSoup
    title = soup.title
    
    productname_h1Tag=soup.find("h1",{"class":"GrE04"})
    Product_Name.append(productname_h1Tag.text)
    
    brandName_tag=soup.find("a",{"context":"brandlabel"})
    brand_name.append(brandName_tag.text)
    
    price_tag=soup.find("div",{"id":"price"})
    price_list.append(price_tag.text)
    
    descrption_tag=soup.find("div",{"id":"about_0"})
    descrpt_list.append(descrption_tag.text)
    
    ImageScarpe(i,soup,productname_h1Tag.text)
```

```python
def ImageScarpe(imgnumb,soup,productname):
    
    # get the image source
    #itemprop="image"
    try:
        div_tag = soup.find_all("div", {"class":"_2FbOx"})[1]
        productimage_tag=div_tag.find("img")
    except:
        div_tag = soup.find_all("div", {"class":"_2FbOx"})[0]
        productimage_tag=div_tag.find("img")
        
    
    #img-responsive bg-logo lazyloaded
   
    # Workbook() takes one, non-optional, argument 
    # which is the filename that we want to create. 
    
    # The workbook object is then used to add new 
    # worksheet via the add_worksheet() method. 
    
    #src = img.get_attribute('src')
    
    #print(image["src"])

    # This is the image url.
    image_url = productimage_tag["src"]
    
    #print(i,image_url)
    # Open the url image, set stream to True, this will return the stream content.
    resp = requests.get(image_url, stream=True)

    # Open a local file with wb ( write binary ) permission.
    #product Image
    local_file = open("D:/Bigbasket Data/Imagescarpe//"+ productname.replace("/"," ") +".jpg", 'wb')
    

    # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
    resp.raw.decode_content = True
    
    #print(local_file)

    # Copy the response stream raw data to local image file.
    shutil.copyfileobj(resp.raw, local_file)
    
    del resp
```
